```python
# Import and download required NLTK data
import nltk

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

text = "Artificial intelligence is changing the world. Google is located in USA."

# Segmentation
sentences = sent_tokenize(text)
print("Segmentation:", sentences)

# Tokenization
tokens = word_tokenize(text)
print("Tokenization:", tokens)

# Stemming
ps = PorterStemmer()
stemmed = [ps.stem(w) for w in tokens]
print("Stemming:", stemmed)

# Lemmatization
lemm = WordNetLemmatizer()
lemmatized = [lemm.lemmatize(w) for w in tokens]
print("Lemmatization:", lemmatized)

# POS tagging
pos_tags = nltk.pos_tag(tokens)
print("POS Tagging:", pos_tags)

print("Named Entity Recognition: NOT SUPPORTED in this version")
```
